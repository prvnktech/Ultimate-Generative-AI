```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import PIL.Image
from IPython.display import display

# We MUST use the compat.v1 bridge because BigGAN is an old format
import tensorflow.compat.v1 as tf1
tf1.disable_eager_execution()

print("Setup Complete.")
```

```python
# Force the loader to treat this as a Legacy Module
model_url = 'https://tfhub.dev/deepmind/biggan-deep-256/1'

print("Loading BigGAN (Legacy Mode)...")
# We use the hub.Module via the tf1 bridge
module = hub.Module(model_url)

# Define the inputs in the Graph
z_placeholder = tf1.placeholder(tf.float32, shape=[None, 128])
y_placeholder = tf1.placeholder(tf.float32, shape=[None, 1000])
truncation_placeholder = tf1.placeholder(tf.float32)

# Create the output tensor
output_tensor = module(dict(z=z_placeholder, y=y_placeholder, truncation=truncation_placeholder))

# Initialize the session
sess = tf1.Session()
sess.run(tf1.global_variables_initializer())
print("Model loaded and Session initialized!")
```

```python
def generate_vehicle(class_id):
    # 1. Create one-hot label
    y_one_hot = np.zeros((1, 1000))
    y_one_hot[0, class_id] = 1

    # 2. Random Noise
    z_noise = np.random.normal(size=(1, 128))

    # 3. Run Session
    samples = sess.run(output_tensor, feed_dict={
        z_placeholder: z_noise,
        y_placeholder: y_one_hot,
        truncation_placeholder: 0.5
    })

    # 4. Post-process
    samples = (samples + 1.0) / 2.0
    img_array = (samples[0] * 255).astype(np.uint8)
    return PIL.Image.fromarray(img_array)
```

```python
# 817 = Sports Car
display(generate_vehicle(817))
```
