```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style for better visuals
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
# Cell 2: Data Generation/Loading
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [15000, 18000, 14500, 22000, 26000, 24500],
    'Expenses': [10000, 11000, 10500, 12000, 13000, 12500]
}

df = pd.DataFrame(data)
df['Profit'] = df['Revenue'] - df['Expenses']

# Displaying the dataframe (Jupyter renders this as a clean table)
df
```

```python
# Cell 3: Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Revenue'], marker='o', label='Revenue', color='teal')
plt.plot(df['Month'], df['Profit'], marker='s', ls='--', label='Profit', color='orange')

plt.title('Monthly Financial Overview')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```
